```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from mc_esoft import MCAgent, MCEpisode
from td_sarsa import SarsaAgent, SarsaEpisode
from td_q_learning import QLearningAgent, QLearningEpisode
```

```python
num_runs = 2500
num_episodes = 1500
```

```python
def run_episode(EpisodeClass, AgentClass):
    rl_agent = AgentClass()
    episode_results = []
    for session in range(num_episodes):
        result = EpisodeClass(rl_agent).play() 
        episode_results.append(result if result==1 else 0)
    return episode_results
```

```python
mc_esoft_result_matrix = []
sarsa_result_matrix = []
qlearning_result_matrix = []

for run in range(num_runs):
    mc_esoft_result_matrix.append(run_episode(MCEpisode,MCAgent))
    sarsa_result_matrix.append(run_episode(SarsaEpisode, SarsaAgent))
    qlearning_result_matrix.append(run_episode(QLearningEpisode,QLearningAgent))
    print(f"Completed {run}")
```

```python
mc_esoft_to_plot = np.array(mc_esoft_result_matrix).mean(axis=0)*100
sarsa_to_plot = np.array(sarsa_result_matrix).mean(axis=0)*100
qlearning_to_plot = np.array(qlearning_result_matrix).mean(axis=0)*100
```

```python
plt.figure(figsize=(10, 10))
plt.title("Successful Solution Percentage vs Episodes")
plt.xlabel("Episodes")
plt.ylabel("% success solution")
plt.plot(mc_esoft_to_plot, label="MC ESoft")
plt.plot(sarsa_to_plot, label="TD SARSA")
plt.plot(qlearning_to_plot, label="TD QLearning")
plt.legend()
```
