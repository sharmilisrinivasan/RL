```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from agent import Agent
from episode import Episode
```

```python
num_runs = 10000
num_episodes = 2000
```

```python
result_matrix = []
for run in range(num_runs):
    rl_agent = Agent()
    episode_results = []
    for session in range(num_episodes):
        result = Episode(rl_agent).play() 
        episode_results.append(1 if result==1 else 0)
    result_matrix.append(episode_results)
    print(f"Completed {run}")
```

```python
len(result_matrix)
```

```python
to_plot = np.array(result_matrix).mean(axis=0)*100
```

```python
plt.figure(figsize=(10, 20))
plt.plot(to_plot) # , legend="MC"
plt.title("Monte Carlo On-policy - Optimal Action Percentage vs Episodes")
plt.xlabel('Episodes')
plt.ylabel("% optimal action")
# plt.legend()
```
